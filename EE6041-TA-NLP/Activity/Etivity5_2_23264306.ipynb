Write a NaiveBayes function which takes the sample training and testing documents shown in the table below as input, and uses the Naive Bayes algorithm to classify the test documents.
Hint1: do not forget to normalise (case-fold) your train & test docs.
Hint2: Python Dictionaries are suitable data structures for storing BOWs (Bag Of Words)

```python
trainingSet = [('London is the Capital of GB','GB'),('Oxford is a city in GB','GB'),('Dublin is the capital of Ireland','IE'),('Limerick is a city in Ireland','IE')]
testSet = [('University of Limerick','?'),('University College Dublin','?'),('Imperial College London','?'),('University of Oxford','?'),('Ireland & GB','?')]
```

```python
def makeMegaDoc(tset, c):
  docsOfc = []
  megaDoc = []
  bow = dict()
  for doc in tset:
    if doc[1] == c:
      docsOfc.append(doc[0])


  for doc in docsOfc:
    wList = doc.split(' ')
    for w in wList:
      if w in bow:
        bow[w] = bow.get(w) + 1
      else:
        bow[w] = 1
      megaDoc.append(w)

  return docsOfc, megaDoc,bow, len(wList)
```

```python
def naiveBayesClassifier(train, test):

  docsOfGB, megaDocGB, GB_BOW, countOfWsGB = makeMegaDoc(train,'GB')
  docsOfIE, megaDocIE, IE_BOW, countOfWsIE = makeMegaDoc(train,'IE')
  probGB = len(docsOfGB)/len(train)
  probIE = len(docsOfIE)/len(train)

  print('MegaDocGB:',megaDocGB)
  print('MegaDocIE:',megaDocIE)

  print('probGB:', probGB)
  print('probIE:', probIE)

  print('GB_BOW', GB_BOW)
  print('IE_BOW', IE_BOW)

  Vcount = sum(GB_BOW.values()) + sum(IE_BOW.values())
  print('V', {**GB_BOW, **IE_BOW})
  print('|V|', Vcount)


  for doc in test:
    print('-----------------------------------------------')
    # for every doc
    print("Text Document:", doc[0])
    docProbGB = probGB
    docProbIE = probIE

    wList = doc[0].split(' ')
    for w in wList:
      # for every words in doc
      cprobGB = 0
      cprobIE = 0
      if w in GB_BOW:
        cprobGB = (GB_BOW[w] + 1)/ (countOfWsGB + Vcount)
      if w in IE_BOW:
        cprobIE = (IE_BOW[w] + 1)/ (countOfWsIE + Vcount)

      docProbGB = float(docProbGB * cprobGB)
      docProbIE = float(docProbIE * cprobIE)

    print("docProbGB:", docProbGB)
    print("docProbIE:", docProbIE)

  if docProbGB > docProbIE:
      print("Inferred Class:GB" )
  else:
      print("Inferred Class:IE" )
```

```python
naiveBayesClassifier(trainingSet,testSet)
```
