### 2.3

```python
def stem_types(types):
  post_stem = set()
  for key in types:
    post_stem.add(ps.stem(key))
  return post_stem
```

```python
def lemmatize_types(types):
  post_lemmatize = set()
  for key in types:
    post_lemmatize.add(lemmatizer.lemmatize(key))
  return post_lemmatize
```

```python
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
post_stem = stem_types(tokenizer.word_counts.keys())
print("Total number of types after stemming: ", len(post_stem))
post_lemmatize = lemmatize_types(tokenizer.word_counts.keys())
print("Total number of types after lemmatizing: ", len(post_lemmatize))
print("total_number_of_types > total_number_of_lemmatized_types > total_number_of_stemmed_types", len(tokenizer.word_counts.keys()) > len(post_lemmatize) > len(post_stem))
```

```python
import spacy
last_100_lines = ShakespeareContent[-100:]
text = ' '.join(last_100_lines)
nlp = spacy.load("en_core_web_sm")
sentences = nlp(text)
assert sentences.has_annotation("SENT_START")
segmented_sents = []
for sent in sentences.sents:
    segmented_sents.append(sent)
print("Segmented Sentences: ", segmented_sents)
print("No. of Segmented Sentences: ", len(segmented_sents))
```
