```python
pos_words = pd.read_csv('positive-words.txt', header=None, names=['Pos_Words'], on_bad_lines='skip',encoding='Latin-1')
neg_words = pd.read_csv('negative-words.txt', header=None, names=['Neg_Words'], on_bad_lines='skip',encoding='Latin-1')
```

```python
neg_words.head(60)
```

```python

no_pos_words = pos_words.shape[0]
no_neg_words = neg_words.shape[0]
```

```python
def sentimentAnalyzer(sentence):
  print("String: " + sentence)
  words = sentence.split(' ')
  pos_count = 0
  neg_count = 0
  for w in words:
    if w in pos_words['Pos_Words'].values:
      pos_count += 1
    elif w in neg_words['Neg_Words'].values:
      neg_count += 1

  if pos_count < neg_count:
    print("Negative Sentiment")
    print("Confidence:", (neg_count / len(words)) )
  elif pos_count > neg_count:
    print("Positive Sentiment")
    print("Confidence:", (pos_count / len(words)) )
  else:
    print("Neutral Sentiment")
```

```python
sentimentAnalyzer("NLP is cool")
```
