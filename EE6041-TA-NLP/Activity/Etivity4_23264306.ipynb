```python
unigrams_df = pd.read_csv('count_1w.txt','\t')
unigrams_df.columns = ['unigram','count']
display(unigrams_df.head(100))
```

### B - Correct Spelling

```python
sum_of_frequecies = sum(unigrams_df['count'])
def nonWordSpellingCorrection(nonWord):
  """
    This function coverts a mis-spelled word into a
    dataframe of possible candidates for the correct intended word

    The output dataframe is sorted automatically as the function
    traverses row by row on a already sorted unigram input file.

    Input: Mis-spelled word (str)
    output: possible candidates (object: pd.DataFrame)

    output columns:
    -> word: the candidate word
    -> Frequency: unigram count or frequency of the word
    -> P(word): Calculated probability of the word appearing in a text

    P(word) = frequency_of_the_word / total_frequency

    return output dataframe
    """

  maxDistance = 2
  resultList = []
  for index, row in unigrams_df.iterrows():
    w = str(row['unigram'])
    minDistance = jellyfish.damerau_levenshtein_distance(nonWord, w)
    if minDistance <= maxDistance:
      count_w = row['count']
      resultList.append([w,count_w, count_w / sum_of_frequecies])

  result_df = pd.DataFrame(resultList, columns=['word', 'frequency','P(w)'])
  return result_df
```

```python
nonWordSpellingCorrection('acress')
```

# Task 2

```python
from textblob import TextBlob
from textblob import Word
```

```python
def getWordConfidence(word):
  """
  Uses the textblob library to create
  possible candidates for the mis-spelled word

  input: mis-spelled word (str)
  output: prints the list of (word, confidence) tuples with spelling suggestions
  """
  w = Word(word)
  return w.spellcheck()
```

```python
print(getWordConfidence('acress'))
```

```python
def correctSpelling(sentence):
  """
  Uses textblob library to correct spellings
  in a sentence

  input: sentence (str)
  output: prints the corrected sentence
  """
  b = TextBlob(sentence)
  return b.correct()
```
