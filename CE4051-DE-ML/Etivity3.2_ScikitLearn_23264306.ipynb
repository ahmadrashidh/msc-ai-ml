## Task 2: Scikit-learn’s Perceptron algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Read the data

1.	Read the iris dataset using Pandas
2.	Extract the Iris-setosa and Iris-verginica samples only
3.	Extract the petal length and petal width. These shall be our features.

```python

df = pd.read_csv('iris_data.csv', header=0)

## Extracting petal length and petal width as features
X = df.iloc[:,2:4].values


## Extracting samples of iris-setosa and iris-verginica
y = df.iloc[:,4].values

## Transforming values to 1 and -1 for Perceptron
y = np.where(y == "Iris-setosa", 1, -1)
```

4.	Plot the dataset as a scatter plot with the two Iris types shown as different colours

```python
plt.scatter(X[:,0], X[:,1],c=y)
```

5.	Use the perceptron in sklearn.linear_model to train on the Iris data previously used. Set verbose = 1; have a look at the epochs, it converges very quickly?

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron as Perceptron

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# set verbose 1 and learning rate to 0.1
model = Perceptron(verbose=1, eta0=0.1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy Score: {accuracy}")
```

6.	Using many random inputs (e.g. 5000) plot to see if you can see the decision boundary for the feature scaled case

```python
no_of_samples = 5000
random_petal_length = np.random.uniform(min(df[df.columns[2]]), max(df[df.columns[2]]), no_of_samples)
random_petal_width = np.random.uniform(min(df[df.columns[3]]), max(df[df.columns[3]]), no_of_samples)
random_data = pd.DataFrame({'petal_length': random_petal_length, 'petal_width': random_petal_width})

predicted_class = model.predict(random_data)

plt.scatter(random_data["petal_length"], random_data["petal_width"],c=predicted_class)
```
